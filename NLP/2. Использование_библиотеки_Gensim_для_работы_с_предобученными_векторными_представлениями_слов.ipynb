# Использование библиотеки Gensim для работы с предобученными векторными представлениями слов

## Введение

Современные технологии обработки естественного языка (NLP) в значительной степени зависят от представления слов как многомерных числовых векторов. Эти вектора, известные как **векторные представления слов** или **word embeddings**, позволяют моделям "понимать" семантические связи между словами. Например, векторное представление слова "король" будет иметь схожесть с вектором слова "королева", так как эти слова имеют похожие значения, относящиеся к королевской власти.

Векторные представления слов используются в различных областях, таких как:
- Классификация текстов,
- Семантический поиск,
- Анализ настроений,
- Построение рекомендательных систем,
- Многочисленные другие задачи, где требуется понимание смысла текста.

Примеры популярных методов для создания word embeddings:
- **Word2Vec**: модель, которая обучается на контекстах слов в тексте.
- **GloVe (Global Vectors for Word Representation)**: метод, который создает векторы на основе глобальной статистики текстов.
- **FastText**: метод, который работает не только с целыми словами, но и с подсловами, что полезно для обработки редких слов и морфологических изменений.

3. Основные операции с векторными представлениями слов.

### Импорт необходимых модулей

Для загрузки модели GloVe мы используем модуль `gensim.downloader`, который предоставляет легкий доступ к популярным предобученным моделям. Импортируем его следующим образом:

### Что такое GloVe?

**GloVe (Global Vectors for Word Representation)** — это одна из самых известных моделей для создания векторных представлений слов. GloVe обучается на большом объеме текстов, таких как:
- Википедия,
- Гига-словари новостей,
- Другие текстовые корпуса.

Основная идея GloVe заключается в том, что каждое слово представляется как точка в многомерном пространстве. Чем более семантически близки слова, тем ближе они расположены в этом пространстве. Например, слова "король" и "королева" будут иметь схожие векторные представления, поскольку они часто появляются в схожих контекстах.

### Загрузка модели

Теперь загрузим предобученную модель GloVe. Мы используем модель с размерностью вектора 50 (50 измерений в векторном пространстве). Эта модель обучена на англоязычной Википедии и корпусе новостей (GloVe Wiki + Gigaword 50). Для этого используем команду:

```python
glove_model = api.load("glove-wiki-gigaword-50")
```

Теперь у нас есть доступ к модели GloVe, и мы можем выполнять различные операции с векторными представлениями слов.

## 3. Основные операции с векторными представлениями

Скачанная модель GloVe позволяет выполнять различные операции с векторами слов. Рассмотрим несколько примеров, таких как извлечение векторов для слов, поиск схожих слов и выполнение арифметических операций.

### 3.1 Получение вектора для слова

Чтобы получить векторное представление для конкретного слова, достаточно просто передать его в модель. Например, чтобы получить вектор для слова `king`, используем следующий код:

```python
word_vector = glove_model['king']
print("Вектор для слова 'king':", word_vector)
```

Вектор будет представлять собой массив чисел, который отображает слово в многомерном пространстве. Этот вектор можно использовать для дальнейших операций, таких как поиск схожих слов или выполнение арифметических операций.

### 3.2 Поиск похожих слов

Модель GloVe позволяет находить слова, которые имеют схожие векторные представления, то есть те слова, которые близки по смыслу. Для этого используется метод `most_similar`, который возвращает список слов, наиболее похожих на заданное. Например, чтобы найти слова, похожие на "king", можно использовать следующий код:

```python
similar_words = glove_model.most_similar('king', topn=5)
print("Слова, похожие на 'king':", similar_words)
```

Результат может быть таким:

```
Слова, похожие на 'king': [('queen', 0.769), ('prince', 0.758), ('monarch', 0.751), ('throne', 0.734), ('kingdom', 0.711)]
```

Здесь числа рядом с каждым словом — это значения схожести, которые показывают, насколько близки слова друг к другу в многомерном пространстве.

### 3.3 Арифметические операции с векторами

Один из уникальных аспектов word embeddings — это возможность выполнения арифметических операций над векторами. Например, можно вычислить вектор, который будет близким к значению "queen", выполнив операцию `woman` + `king` - `man`. Это работает благодаря тому, что модель GloVe "понимает" семантические связи между словами.

Пример:

```python
result = glove_model.most_similar(positive=['woman', 'king'], negative=['man'], topn=5)
print("Результат операции 'woman' + 'king' - 'man':", result)
```

Ожидаемый результат:

```
Результат операции 'woman' + 'king' - 'man': [('queen', 0.831), ('monarch', 0.784), ('princess', 0.762), ('throne', 0.753), ('empress', 0.742)]
```

Этот пример иллюстрирует, как word embeddings могут моделировать не только синтаксические, но и семантические отношения между словами.

## 4. Что такое предобученная модель?

**Предобученные модели** — это модели машинного обучения, которые были обучены на большом наборе данных и могут быть использованы для решения задач без необходимости повторного обучения с нуля. В контексте векторных представлений слов предобученные модели, такие как GloVe, Word2Vec или FastText, обучаются на огромных корпусах текстов и сохраняют векторы для слов. Эти модели позволяют использовать их напрямую для анализа текстов, поиска схожих слов или выполнения других задач без необходимости собирать данные и обучать модель заново. Это значительно ускоряет процесс разработки и позволяет сэкономить ресурсы.

Предобученные модели могут быть адаптированы под конкретные задачи с помощью дополнительного обучения (fine-tuning), если этого требует специфика задачи.

## 5. Полный пример


````python
# Установка необходимых библиотек
!pip install gensim

# Импортируем библиотеку
import gensim.downloader as api

# Загрузка модели GloVe
glove_model = api.load("glove-wiki-gigaword-50")

# Получение вектора для слова
word_vector = glove_model['king']
print("Вектор для слова 'king':", word_vector)

# Поиск слов, наиболее похожих на 'king'
similar_words = glove_model.most_similar('king', topn=5)
print("Слова, похожие на 'king':", similar_words)

# Операции с векторами
result = glove_model.most_similar(positive=['woman', 'king'], negative=['man'], topn=5)
print("Результат операции 'woman' + 'king' - 'man':", result)
````

Визуализация векторных представлений слов позволяет понять их взаимосвязи и группировку в пространстве. Поскольку векторные представления часто имеют высокую размерность (например, 50, 100 или 300), их необходимо сначала преобразовать в двумерное или трехмерное пространство. Для этого можно использовать такие методы, как t-SNE или PCA.

Ниже приведен пример кода, который показывает, как визуализировать векторы с использованием библиотеки matplotlib.



````python
# Импортируем библиотеки
import gensim.downloader as api
import plotly.graph_objects as go
from sklearn.manifold import TSNE
import numpy as np

# Загрузка предобученной модели GloVe
glove_model = api.load("glove-wiki-gigaword-50")

# Список слов для визуализации
words = ['king', 'queen', 'man', 'woman', 'prince', 'princess',
         'throne', 'monarch', 'kingdom', 'duke', 'duchess',
         'lord', 'lady', 'emperor', 'empress']

# Получение векторов для выбранных слов
word_vectors = np.array([glove_model[word] for word in words])

# Преобразование векторов в 3D-пространство с помощью t-SNE
tsne = TSNE(n_components=3, random_state=42, perplexity=10, n_iter=1000)
word_vectors_3d = tsne.fit_transform(word_vectors)

# Создание интерактивного 3D-графика
fig = go.Figure()

# Добавление точек на график
fig.add_trace(
    go.Scatter3d(
        x=word_vectors_3d[:, 0],  # Координата X
        y=word_vectors_3d[:, 1],  # Координата Y
        z=word_vectors_3d[:, 2],  # Координата Z
        mode='markers+text',      # Режим: точки + текст
        text=words,               # Подписи точек
        marker=dict(size=8, color='blue', opacity=0.8),
        textposition="top center"
    )
)

# Настройка отображения графика
fig.update_layout(
    title="Интерактивная 3D-визуализация векторных представлений слов",
    scene=dict(
        xaxis_title="Компонента 1",
        yaxis_title="Компонента 2",
        zaxis_title="Компонента 3",
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

# Показ графика
fig.show()
````


### Дополнительно: Использование PCA вместо t-SNE

Если t-SNE слишком медленный или данные очень большие, можно использовать PCA (Principal Component Analysis) для уменьшения размерности:


````python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Преобразование векторов в 3D-пространство с помощью PCA
pca = PCA(n_components=3)
word_vectors_3d_pca = pca.fit_transform(word_vectors)

# Построение 3D-графика с использованием matplotlib
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Координаты точек
x = word_vectors_3d_pca[:, 0]
y = word_vectors_3d_pca[:, 1]
z = word_vectors_3d_pca[:, 2]

# Построение точек
ax.scatter(x, y, z, color='blue', s=50)

# Добавление подписей слов
for i, word in enumerate(words):
    ax.text(x[i], y[i], z[i], word, fontsize=10, color='red')

# Настройка графика
ax.set_title("Визуализация векторных представлений слов (PCA, 3D)", fontsize=16)
ax.set_xlabel("PCA компонент 1", fontsize=12)
ax.set_ylabel("PCA компонент 2", fontsize=12)
ax.set_zlabel("PCA компонент 3", fontsize=12)
plt.show()
````


Очень классическим примером этого является то, что если вы возьмете разницу между векторами для "женщины" и "мужчины" — это можно визуализировать как маленький вектор, соединяющий концы одного вектора с концом другого — то эта разница будет очень похожа на разницу между "королем" и "королевой".

Предположим, вы не знаете слова для обозначения female monarch. Вы могли бы найти его, взяв вектор для "короля", добавив направление от "мужчины" к "женщине" и найдя векторы, ближайшие к этой точке. По крайней мере, это идея. Несмотря на то, что это классический пример для модели, с которой я работаю, фактический вектор слова "королева" оказывается немного дальше, чем это предположение. Это несоответствие, вероятно, связано с тем, что слово "королева" используется в обучающих данных не просто как женская версия "короля".

Когда я экспериментировал с моделью, семейные отношения казались гораздо более наглядными для иллюстрации этой концепции. Главное, что стоит отметить, это то, что во время обучения модель нашла целесообразным выбирать векторы таким образом, чтобы одно направление в этом пространстве кодировало информацию о поле.

Еще один интересный пример: если взять вектор для "Италии", вычесть вектор для "Германии" и добавить вектор для "Гитлера", то вы получите вектор, очень близкий к вектору "Муссолини".


````python
import gensim.downloader as api
from gensim.models import KeyedVectors

# Загрузка предобученной модели Word2Vec
model = api.load("word2vec-google-news-300")

# Пример векторов
woman_vector = model['woman']
man_vector = model['man']
king_vector = model['king']
queen_vector = model['queen']

# Вычисление разницы между векторами
gender_difference = woman_vector - man_vector
# Находим вектор для female monarch
female_monarch_vector = king_vector + gender_difference

# Находим ближайшее слово к полученному вектору
similar_words = model.similar_by_vector(female_monarch_vector, topn=5)

# Выводим результаты
print("Ближайшие слова к вектору female monarch:")
for word, similarity in similar_words:
    print(f"{word}: {similarity:.4f}")

# Пример с Италией и Германией
italy_vector = model['Italy']
germany_vector = model['Germany']
hitler_vector = model['Hitler']

# Вычисляем вектор для Муссолини
mussolini_vector = italy_vector - germany_vector + hitler_vector
similar_mussolini = model.similar_by_vector(mussolini_vector, topn=5)

# Выводим результаты
print("\nБлижайшие слова к вектору для Муссолини:")
for word, similarity in similar_mussolini:
    print(f"{word}: {similarity:.4f}")
````

````python
# Пример векторов
woman_vector = model['woman']
man_vector = model['man']
king_vector = model['king']
queen_vector = model['queen']

# Функция для вычисления скалярного произведения
def scalar_product(vec1, vec2):
    return np.dot(vec1, vec2)

# Функция для вычисления косинусного расстояния
def cosine_distance(vec1, vec2):
    return 1 - np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

# Вычисление скалярного произведения
scalar_product_woman_man = scalar_product(woman_vector, man_vector)
scalar_product_king_queen = scalar_product(king_vector, queen_vector)

# Вычисление косинусного расстояния
cosine_distance_woman_man = cosine_distance(woman_vector, man_vector)
cosine_distance_king_queen = cosine_distance(king_vector, queen_vector)

# Вывод результатов
print(f"Скалярное произведение векторов 'женщина' и 'мужчина': {scalar_product_woman_man:.4f}")
print(f"Скалярное произведение векторов 'король' и 'королева': {scalar_product_king_queen:.4f}")
print(f"Косинусное расстояние между векторами 'женщина' и 'мужчина': {cosine_distance_woman_man:.4f}")
print(f"Косинусное расстояние между векторами 'король' и 'королева': {cosine_distance_king_queen:.4f}")
````
